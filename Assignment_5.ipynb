Q3 (a)

```python
first_initial = 'P'
last_initial = 'G'
X = ord(first_initial) + ord(last_initial)
print(X)
sales = np.array([X, X + 50, X + 100, X + 150, X + 200])
print("Initial Sales Data:", sales)
```

(b)

```python
tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales + (sales * tax_rate)
taxed_sales
```

(c)

```python
discounts = np.where(sales < (X + 100), 0.05, 0.10)
adjusted_sales = sales - (sales * discounts)
adjusted_sales
```

(d)

```python
random_sales = np.random.randint(sales.min(), sales.max()+1, size=(2,len(sales)))
weekly_sales = np.vstack((sales,random_sales))
weekly_sales
```

```python
growth_rates = np.array([1, 1.02, 1.04]).reshape(-1, 1)
weekly_sales = weekly_sales * growth_rates
weekly_sales
```

Q4

```python
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
```

```python
y1 = x**2
plt.plot(x, y1)
plt.title('Graph of X^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
y2 = np.sin(x)
plt.plot(x, y2)
plt.title('Graph of sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
y3 = np.exp(x)
plt.plot(x, y3)
plt.title('Graph of e^x')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
y4 = np.log(np.abs(x)+1)
plt.plot(x, y4)
plt.title('Graph of log(|x| + 1)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
