```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
roll_no =  102317214
np.random.seed(roll_no)
```

```python
sales_data = np.random.randint(1000, 5000, size=(12, 4))
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)
df
```

```python
print("First 5 rows of DataFrame:\n", df.head())
print("\nSummary Statistics:\n", df.describe())
```

```python
total_sales_category = df.sum()
print("\nTotal Sales per Category:\n", total_sales_category)
```

```python
df['Total Sales'] = df.sum(axis=1)
df
```

```python
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
df
```

```python
df['Electronics'] = df['Electronics'] * 0.9
df
```

```python
plt.figure(figsize=(10, 6))
for category in categories:
    plt.plot(df.index, df[category], label=category)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Trends')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.show()
```

```python
def swap_list_elements(lst, pos1, pos2):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp
    return lst

lst = [1, 2, 3, 4, 5]
print("Swapped List:", swap_list_elements(lst, 1, 3))
```

```python
def swap_set_elements(s, pos1, pos2):
    s_list = list(s)
    s_list[pos1], s_list[pos2] = s_list[pos2], s_list[pos1]
    return set(s_list)

s = {10, 20, 30, 40, 50}
print("Swapped Set:", swap_set_elements(s, 0, 3))
```
