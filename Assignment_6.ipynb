```python
import matplotlib.pyplot as plt
import numpy as np
M = float(input('Enter the value of M:'))
x = np.linspace(-10,10,1000)
y1 = M*(x**2)
y2 = M*(np.sin(x))
```

```python
plt.plot(x,y1)
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x,y2)
plt.title('Sin Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x, y1, label='M * x^2', color='blue', linestyle='-')
plt.plot(x, y2, label='M * sin(x)', color='red', linestyle='--')
plt.title('Quadratic and Sine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import random
data = {
    'Subject':['Math','English','Hindi','Punjabi','Science']
}
df = pd.DataFrame(data)
df['Marks'] = [random.randint(10,100) for i in range(5)]
df
```

```python
ax = sns.barplot(x='Subject', y='Marks', data=df, palette='coolwarm')
for i, value in enumerate(df['Marks']):
    ax.text(i, value+1, str(value), ha='center', fontsize=9)
plt.title('Student Marks by Subject', fontsize=14)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Marks', fontsize=12)
plt.grid(axis='y')
plt.show()
```

```python
roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)
cumulative_sum = np.cumsum(data)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(cumulative_sum, color='blue', linestyle='-', marker='o', markersize=4)
axes[0, 0].set_title('Cumulative Sum')
axes[0, 0].set_xlabel('Index')
axes[0, 0].set_ylabel('Cumulative Sum')
axes[0, 0].grid(True)

axes[0, 1].scatter(range(50), data, color='red', marker='x')
axes[0, 1].set_title('Scatter Plot (Random Noise)')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Value')
axes[0, 1].grid(True)

axes[1, 0].axis('off')
axes[1, 1].axis('off')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('company_sales_data.csv')
df
```

```python
df['total_profit']
```

```python
plt.plot(df['month_number'], df['total_profit'], marker='o', linestyle='-', color='b')
plt.title('Total Profit by Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
```

```python
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo','moisturizer']
for product in products:
  sns.lineplot(data=df,x='month_number',y=product,label=product)
plt.title('Product Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
```

```python
df_melted = df.melt(
    id_vars=['month_number'],
    value_vars=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'],
    var_name='Product',
    value_name='Sales'
)
sns.barplot(data=df_melted, x='Product', y='Sales', hue='month_number')
plt.title('Product Sales by Month')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend(title='Month')
plt.show()
```
